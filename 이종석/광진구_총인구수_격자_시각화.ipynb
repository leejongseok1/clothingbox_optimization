```python
import pandas as pd
import geopandas as gpd
import folium
```

```python
geo_data = gpd.read_file('/content/drive/MyDrive/GWANGJIN/Grid/광진구_총인구수(202310)_격자.geojson')
hubo = gpd.read_file('/content/drive/MyDrive/GWANGJIN/광진구_의류수거함_위치좌표.gpkg')
```

```python
center = [37.5467284, 127.0857543]; zoom = 14 # 광진구 centroid 좌표

m = folium.Map(
    location=center,
    zoom_start=zoom,
)

map = folium.Choropleth(
    geo_data=geo_data,
    data=geo_data,
    columns=['gid', 'val'],
    key_on='feature.properties.gid',
    fill_color="YlGn",
    legend_name="population"
    # ,point=hubo
).add_to(m)

folium.GeoJsonPopup(fields=['val']).add_to(map.geojson)

m
```

```python
hubo = hubo[['No','Address','geometry']]
hubo['geometry'] = gpd.GeoSeries(hubo.buffer(100))
hubo = hubo.to_crs(4326)
```
