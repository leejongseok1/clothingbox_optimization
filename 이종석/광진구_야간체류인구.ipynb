```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/GWANGJIN/LOCAL_PEOPLE_DONG_202403.csv')
mapp = pd.read_excel('/content/drive/MyDrive/GWANGJIN/행정동코드_매핑정보_20200325.xlsx')[1:]
mapp.rename(columns={'행자부행정동코드':'행정동코드'}, inplace=True)
mapp['행정동코드'] = pd.to_numeric(mapp['행정동코드'])
mapp.drop(['통계청행정동코드', '시도명'], axis=1, inplace=True)
mapp.reset_index(drop=True, inplace=True)
```

```python
filter = mapp.loc[mapp['시군구명'] == '광진구']

merged = pd.merge(df, filter, how='inner', on='행정동코드')
merged = merged[(merged['시간대구분'] >= 0) & (merged['시간대구분'] <= 6)]
merged
```

```python
merged_mean = merged.groupby('행정동코드')[['총생활인구수']].mean()
merged_mean = pd.merge(merged_mean, mapp[['행정동코드', '행정동명']], on='행정동코드', how='left')
merged_mean.rename(columns={'총생활인구수':'야간체류인구수'}, inplace=True)

merged_mean
```

```python
merged_mean.to_csv('광진구_동별야간체류인구수.csv', index=False)
```
