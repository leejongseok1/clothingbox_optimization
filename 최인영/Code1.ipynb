```python
import pandas as pd
import numpy as np

df = pd.read_csv('LOCAL_PEOPLE_DONG_202402.csv')
mapp = pd.read_excel('행정동코드_매핑정보_20200325 (2).xlsx')[1:]
mapp['행자부행정동코드'] = pd.to_numeric(mapp['행자부행정동코드'])
df.reset_index(inplace=True)
tpcolumns = df.columns[1:]
df.drop(columns='여자70세이상생활인구수', inplace = True)
df.columns = tpcolumns
df

# 2024년 2월의 서울시 생활인구 행정동단위 데이터와, 행정동코드 매핑정보 파일을 활용한다.
# 칼럼 밀리는 현상 수정완료.
```

```python
dddd = mapp.loc[mapp['시군구명'] == '광진구']
df1 = pd.DataFrame()
for i in range(15):
    df1 = df1.append(df.loc[df['행정동코드'] == dddd.행자부행정동코드.tolist()[i]])
    
df1

# 1. 행정동코드 매핑정보 파일에서 '시군구명'이 광진구인 동의 행만 뽑아낸다. 
# 2. 2024년 서울시 생활인구 행정동별 데이터에서 '행정동코드'가 광진구의 행정동인 행만 뽑아낸다. -> 컬럼이 오른쪽으로 한 칸씩 밀려 코드상엔 임시로 '시간대구분'이라 적었으나, 실제로는 '행정동코드'  수정완료
```

```python
df1_groupby = df1.groupby('행정동코드')[['총생활인구수']].sum()
df1_groupby

# 광진구에서 2024년 2월 생활인구가 가장 많았던 동을 찾는 과정이다.
# 역시 컬럼이 하나씩 밀려 코드상엔 '시간대구분', '행정동코드'로 임시 작성하였으나, 실제로는 '행정동코드', '총생활인구수' 이다.  수정완료

df1_groupbyJunggok = df1_groupby[2:6].sum(axis=0)
df1_groupbyJayang = df1_groupby[8:12].sum(axis=0)
df1_groupbyGuui = df1_groupby[12:15].sum(axis=0)

# 법정동으로 통합하려면 위의 코드와 같이 통합하고 데이터프레임에 붙여 넣어준다. 
```

```python
df_tp = pd.concat([df1_groupby.reset_index(),df2], axis = 1)
df_tp

# 광진구 동별 생활인구 데이터와 동별 생활인구 데이터를 통합하여 데이터프레임화 
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df_tp[['총생활인구수','소계']])
df2_scaled = scaler.transform(df_tp[['총생활인구수','소계']])

df2_scaled_d = pd.DataFrame(data=df2_scaled, columns = ['총생활인구수', '소계'])
df2_scaled_d

# 총 생활인구수와 주민등록인구의 차(diff)를 이용하여 주민등록인구 대비 생활인구가 많은 지역, 생활인구 대비 주민등록인구가 많은 지역 탐색
# 데이터의 스케일을 맞춰주기 위해 표준화 작업 시행 (StandardScaler)
```

```python
df3 = df2_scaled_d.diff(axis = 1)
df3

# 주민등록인구 - 총생활인구수의 값이므로, 작은 값일수록 주민등록인구대비 총생활인구수가 많은 지역이다. 
# 반대로 큰 값일수록 총생활인구대비 주민등록인구가 많은 지역이다.
# 해당 지역을 이동이 많은 교통이 활발한 지역으로 판단할 수 있으므로(물론 주관적이라 더 적합한 판단을 찾으면 좋음) 도시미관, 통행에 영향을 끼치는 의류수거함 배치의 재고가 필요하다. 
# 생활인구 자체가 많은 지역도, 지역 내에서 활발한 움직임이 나타난다고 볼 수 있기 때문에, 이런 점의 고려도 필요하다. 
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

train = pd.read_excel('343.xlsx')
train.pop('지역')

y_tr = train.pop('의류폐기물양')
x_train, x_val, y_train, y_val = train_test_split(train, y_tr, test_size = 0.3)
model_RF = RandomForestRegressor(n_estimators = 200, random_state = 2023)
model_RF.fit(x_train, y_train)

pred_test = model_RF.predict([[2022,17.06,351252]])
pred_test

# 임시로 2007년 ~ 2019년 서울 지역의 면적, 의류폐기물양, 인구수 데이터를 활용하여 랜덤포레스트 회귀 예측모델을 만들고,
# 광진구의 2022년 의류폐기물 양을 예측해보는 것을 해보았는데, 2019년까지의 데이터는 2020년 이후 데이터와 의류폐기물 
# 양 차이가 심하고, 데이터가 충분하지 못해 오차가 상당히 큰 것 같습니다. 하지만 모델이 돌아가는지 확인해보기 위함이었으므로 큰 의미는 없습니다.
# 아예 2022년부터 2023년까지의 광진구 의류폐기물 양 데이터로 학습을 진행하는 것이 더 정확할 것 같습니다.
```

```python
import json
import folium
import pandas as pd

geo_path = 'gj.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))

pp = pd.read_excel('a.xlsx')
data1 = pd.read_excel('b.xlsx')

m = folium.Map(location=[37.5383742, 127.0822077], zoom_start=14) #광진구청 위치로 임시로 잡음 
folium.Choropleth(geo_data=geo_str,
            data=pp,
            columns=['행정동코드','a'],
            fill_color='PuRd',
            key_on='properties.code',
            highlight=True,
            fill_opacity=0.5, line_opacity=1,
            legend_name='aaa').add_to(m)

for i, j in zip(data1.Latitude, data1.Longitude):
    folium.CircleMarker([i, j], radius = 3).add_to(m)
    
m

# 전국 동별 경도, 위도 정보가 담긴 json 파일 'https://raw.githubusercontent.com/southkorea/southkorea-maps/master/kostat/2013/json/skorea_submunicipalities_geo_simple.json'
# 위의 데이터에서 광진구의 동만 뽑아서 활용. 'gj.json' 파일
# 'a.xlsx' 파일은 각 동의 행정동코드와 의류수거함수/주민등록인구수 값이 나타난 파일.
# 'b.xlsx' 파일은 의류수거함 각각의 경도, 위도 정보가 담긴 파일.
# 지도가 조금 문제인건지 오른쪽 하단 쪽으로 살짝 밀려 표시되는 느낌인데, 대략적으로 구역별 표현이 가능.
# 색이 진할수록 주민등록인구대비 의류수거함이 많은 지역.
# 이를 이용하여 다른 정보를 활용한 지도 시각화도 가능. 
```

```python
for i, j in zip(data1.Latitude, data1.Longitude):
    folium.CircleMarker([i, j], radius = 3).add_to(m)
    folium.Circle(
    location=[i, j],
    radius=130, # 원 크기
    color='00ff0000', # 원 선 색상
    fill_color='blue', # 원 내부 색상
    popup='Circle popup',
    tooltip='Circle tooltip'
).add_to(m)
    
m

# folium.Circle 옵션을 추가하여 원으로 범위 표현 가능
# 색이 진한 지역은 의류수거함 접근이 쉬운 지역이고, 연한 지역은 접근이 어려운 지역으로 판단 가능 
```
