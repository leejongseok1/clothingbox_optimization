```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

mapp = pd.read_excel('행정동코드_매핑정보_20200325 (2).xlsx')[1:]
mapp['행자부행정동코드'] = pd.to_numeric(mapp['행자부행정동코드'])
df = pd.read_csv('data.csv', encoding='cp949')

# 행정동코드, 생활인구데이터 통합 파일 불러오기 
```

```python
dddd = mapp.loc[mapp['시군구명'] == '광진구']
df1 = pd.DataFrame()
for i in range(15):
    df1 = df1.append(df.loc[df['행정동코드'] == dddd.행자부행정동코드.tolist()[i]])
    
# df1은 생활인구데이터에서 광진구 데이터만을 뽑은 데이터 
```

```python
df1_groupby = df1.groupby('행정동코드')[['총생활인구수']].sum()
df2 = pd.read_excel('주민등록인구_20240401234727.xlsx')
# df1_groupby는 광진구 행정동별 총생활인구수를 나타낸 데이터
# df2는 광진구 동별 주민등록인구데이터 (2023년 4분기 기준)
```

```python
df_tp = pd.concat([df1_groupby.reset_index(),df2], axis = 1)
df_tp
# 광진구 동별 생활인구 데이터와 주민등록인구 데이터를 통합하여 데이터프레임화 
```

```python
dtime = df1.loc[df1['시간대구분'] < 7].append(df1.loc[df1['시간대구분'] > 21])
df3_groupby = dtime.groupby('행정동코드')[['총생활인구수']].sum()

# dtime은 df1에서 야간시간대(오후 22시부터 다음날 오전 6시까지) 데이터만을 뽑은 데이터 
# df3_groupby는 dtime에서 행정동별 총생활인구수를 뽑은 데이터 (거주인구)
```

```python
df_ha = pd.concat([df1_groupby, df3_groupby], axis = 1)
df_ha.columns = ['총생활인구수', '생활인구수_야간']

df_ha['주민등록인구'] = [27645,21465,15691,20882,16008,28006,11496,34210,22745,23949,27543,23405,23833,26230,28072] # df2.소계 이런식으로 하려했는데 자꾸 nan로 떠서 임시로...
df_ha['동'] = ['화양동','군자동','중곡1동','중곡2동','중곡3동','중곡4동','능동','광장동','자양1동','자양2동','자양3동','자양4동','구의1동','구의2동','구의3동']
df_ha['Code'] = [11050530,11050540,11050550,11050560,11050570,11050580,11050590,11050630,11050640,11050650,11050660,11050670,11050600,11050610,11050620]
df_ha
```

```python
scaler = StandardScaler()
scaler.fit(df_ha[['총생활인구수','생활인구수_야간', '주민등록인구']])
df2_scaled = scaler.transform(df_ha[['총생활인구수','생활인구수_야간', '주민등록인구']])

df2_scaled_d = pd.DataFrame(data=df2_scaled, columns = ['총생활인구수','생활인구수_야간', '주민등록인구'])
df2_scaled_d.index = df_ha.index
df2_scaled_d['생활인구수_주간'] = df2_scaled_d.총생활인구수 - df2_scaled_d.생활인구수_야간
df2_scaled_d['생활인구수_주야차'] = df2_scaled_d.생활인구수_주간 - df2_scaled_d.생활인구수_야간
df2_scaled_d['주민등록인구-총생활인구'] = df2_scaled_d.주민등록인구 - df2_scaled_d.총생활인구수
df2_scaled_d[['동','Code']] = df_ha[['동','Code']]
df2_scaled_d

# 데이터의 스케일을 맞춰주기 위해 표준화 작업 시행 (StandardScaler)
```
