```python
import pandas as pd
import numpy as np

df = pd.read_csv('LOCAL_PEOPLE_DONG_202402.csv')
mapp = pd.read_excel('행정동코드_매핑정보_20200325 (2).xlsx')[1:]
mapp['행자부행정동코드'] = pd.to_numeric(mapp['행자부행정동코드'])
df.reset_index(inplace=True)
tpcolumns = df.columns[1:]
df.drop(columns='여자70세이상생활인구수', inplace = True)
df.columns = tpcolumns
df
```

```python
dddd = mapp.loc[mapp['시군구명'] == '광진구']
df1 = pd.DataFrame()
for i in range(15):
    df1 = df1.append(df.loc[df['행정동코드'] == dddd.행자부행정동코드.tolist()[i]])
    
df1
```

```python
dtime = df1.loc[df1['시간대구분'] < 7].append(df1.loc[df1['시간대구분'] > 21])
    
dtime
```

```python
#df1_groupby = dtime.groupby(['행정동코드','시간대구분'])[['총생활인구수']].sum()
#df1_groupby

df1_groupby = dtime.groupby('행정동코드')[['총생활인구수']].sum()
df1_groupby
```

```python
df0_groupby = df1.groupby('행정동코드')[['총생활인구수']].sum()
df0_groupby
```

```python
df_ha = pd.concat([df0_groupby, df1_groupby], axis = 1)
df_ha.columns = ['총생활인구수', '생활인구수_야간']

df_ha['생활인구수_주간'] = df_ha.총생활인구수 - df_ha.생활인구수_야간
df_ha['생활인구수_주야차'] = df_ha.생활인구수_주간 - df_ha.생활인구수_야간
df_ha['동'] = ['화양동','군자동','중곡1동','중곡2동','중곡3동','중곡4동','능동','광장동','자양1동','자양2동','자양3동','자양4동','구의1동','구의2동','구의3동']
df_ha
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc("font", family = "Malgun Gothic")
sns.set(font="Malgun Gothic", 
rc={"axes.unicode_minus":False}, style='white',font_scale=0.6)

#시각화를 위한 라이브러리 임포트, 폰트 설정 

sns.barplot(x= df_ha.동, y= '생활인구수_주간', data = df_ha)
plt.title("광진구 각 동별 생활인구수_주간")
plt.show()
```
